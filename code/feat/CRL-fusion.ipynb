### 三个模块特征融合

```python
a = np.load(r'...\data\piRDieasev1.0\DCNN\dcnn_feat.npy') 

b = np.load(r'...\data\piRDieasev1.0\GCN\gmodel_feat.npy') 

c = np.load(r'...\data\piRDieasev1.0\Trans\trans_feat.npy') 

label = pd.read_csv(r'...\data\piRDieasev1.0\Trans\label_bal.csv',index_col=0)
index = label.index
```
