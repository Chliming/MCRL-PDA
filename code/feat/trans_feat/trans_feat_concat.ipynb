```python
import torch.nn as nn
import torch
import pandas as pd
import numpy as np
```

### 1 拼接

```python
adj = pd.read_csv(r'...\data\piRDieasev1.0\adj.csv',index_col=0)  # 4981
```

```python
trans_feat = pd.DataFrame(np.loadtxt(r'...\data\piRDieasev1.0\Trans\trans_heter.txt')) # 4370 4370

piR = np.array(trans_feat.iloc[:4349])   # 4349 21
dis = np.array(trans_feat.iloc[4349:])
```

```python
ld = []  
for i in range(piR.shape[0]):
    for j in range(dis.shape[0]):
        ld1 = np.hstack((piR[i],dis[j]))
        ld.append(ld1)
```

### 2 正负平衡

```python
feat = np.load(r'...\data\piRDieasev1.0\Trans\trans_heter_ld_label.npy')
feat = pd.DataFrame(feat)

dic_piR = dict(enumerate(adj.index))
dic_dis = dict(enumerate(adj.columns))

index = []
for i in range(adj.shape[0]):
    for j in range(adj.shape[1]):
        index.append(dic_piR[i]+'+'+dic_dis[j])

feat.index = index
```

```python
feat
```

```python
data = pd.concat([feat[feat.iloc[:,-1]==1],
                  feat[feat.iloc[:,-1]==0].sample(frac=1, random_state=42).iloc[:4981,:]],axis=0)
```
